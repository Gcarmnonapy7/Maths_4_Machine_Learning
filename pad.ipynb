```python
from sklearn.cluster import KMeans
from kmodes.kprototypes import KPrototypes
import pandas as pd

df = pd.DataFrame({
    "price": [10, 12, 50, 55],
    "rooms": [3, 4, 10, 12],
    "distributor": ["A", "A", "B", "B"]
})

X = df[["price"]]

kproto = KPrototypes(n_clusters=2, init='Cao', verbose=1)
categorical_cols = [2]
clusters = kproto.fit_predict(df, categorical=categorical_cols)

df["cluster"] = clusters
```
